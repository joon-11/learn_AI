```python
import os
import cv2
import numpy as np

dirs = ['ants', 'bees']
pixels = []
labels = []

for i, d in enumerate(dirs):
    files = os.listdir('./data/' + d)
    for f in files:
        img = cv2.imread("./data/" + d + "/" + f, 0)
        img = cv2.resize(img, (128, 128))
        img = np.array(img).flatten().tolist()#flattern = 2차원 배열 1차원 배열로 바꿔줌,
        pixels.append(img)
        labels.append(i)
```

```python
import pandas as pd

pixels_df = pd.DataFrame(pixels)
pixels_df = pixels_df/255 #픽셀값을 0 과 1 사이에 들어갈수 있도록 정규화

labels_df = pd.DataFrame(labels)
labels_df = labels_df.rename(columns = {0: 'label'})

img_set = pd.concat([pixels_df, labels_df], axis = 1)
img_set.head()
```

```python
import os
import cv2
import numpy as np

dirs = ['ants', 'bees']
pixels_color = []
labels_color = []

for i, d in enumerate(dirs):
    files = os.listdir('./data/' + d)
    for f in files:
        img = cv2.imread("./data/" + d + "/" + f)
        img = cv2.resize(img, (128, 128))
        b, g, r = cv2.split(img)
        b = np.array(b).flatten().tolist()
        g = np.array(g).flatten().tolist()
        r = np.array(r).flatten().tolist()
        tmp = b + g + r
        pixels_color.append(tmp)
        labels_color.append(i)
```

```python
import pandas as pd

pixels_df2 = pd.DataFrame(pixels_color)
pixels_df2 = pixels_df2/255 #픽셀값을 0 과 1 사이에 들어갈수 있도록 정규화

labels_df2 = pd.DataFrame(labels_color)
labels_df2 = labels_df.rename(columns = {0: 'label'})

img_set_color = pd.concat([pixels_df2, labels_df2], axis = 1)
img_set_color.head()
```

**모폴로지 변환**

이진화 이미지를 대상으로 압축, 팽창, 오프닝, 클로징 등을 처리

침식(erosion)
커널 영역 안에 존재하는 모든 픽셀의 값을 커널 내부의 극솟값(local minimum)으로 대체합니다.
즉, 구조 요소(element)를 활용해 이웃한 픽셀을 최소 픽셀값으로 대체합니다.
침식 연산을 적용하면 밝은 영역이 줄어들고 어두운 영역이 늘어납니다.
침식 연산은 노이즈 제거에 주로 사용합니다.

팽창(dilation)
커널 영역 안에 존재하는 모든 픽셀의 값을 커널 내부의 극댓값(local maximum)으로 대체합니다.
즉, 구조 요소(element)를 활용해 이웃한 픽셀들을 최대 픽셀값으로 대체합니다.
팽창 연산을 적용하면 어두운 영역이 줄어들고 밝은 영역이 늘어납니다.
팽창 연산은 노이즈 제거 후 줄어든 크기를 복구하고자 할 때 주로 사용합니다.

오프닝
침식을 먼저 하고 팽창을 하는 알고리즘
침식을 사용하게 되면 불필요한 요소도 떨어지지만 사이즈 자체도 작아지는 단점이 발생함.

클로징
팽창을 먼저 하고 침식을 하는 알고리즘

```python
## 침식

import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('./data/ants/swiss-army-ant.jpg', 0)
ret, bin_img = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY)

kernel = np.ones((3,3), np.uint8)
img_el = cv2.erode(bin_img, kernel, iterations = 1)
plt.imshow(img_el, cmap = 'gray')
```

```python
## 팽창

img_dl = cv2.dilate(bin_img, kernel, iterations = 1)
plt.imshow(img_dl, cmap = 'gray')
```

```python
##오프닝

img_op = cv2.morphologyEx(bin_img, cv2.MORPH_OPEN, kernel)
plt.imshow(img_op, cmap = 'gray')
```

```python
## 클로징

img_cl = cv2.morphologyEx(bin_img, cv2.MORPH_CLOSE, kernel)
plt.imshow(img_cl, cmap = 'gray')
```

즉, 쓸모 없는 데이터가 너무 많아서 학습이 되지 않는 현상을 이러나지 않게 하려고
고차원 데이터를 저차원 데이터로 변환해주는 알고리즘 이다.

**이미지**

가지고 있는 데이터 수가 적을땐 이미지를 반전시키거나 블러처리, 명도변경을 함으로써 데이터 수를 늘릴수 있음.

```python
img = cv2.imread('./data/ants/swiss-army-ant.jpg', 0)

x_img = cv2.flip(img,0)
y_img = cv2.flip(img,1)
xy_img = cv2.flip(img,-1)

cv2.imwrite('x_img.jpg', x_img)
cv2.imwrite('y_img.jpg', y_img)
cv2.imwrite('xy_img.jpg', xy_img)
```

```python
blur_img = cv2.blur(img, (5,5))
gau_img = cv2.GaussianBlur(img,(5,5),0)
med_img = cv2.medianBlur(img,3)

cv2.imwrite('blue_img.jpg', blur_img)
cv2.imwrite('gau_img.jpg', gau_img)
cv2.imwrite('med_img.jpg', med_img)
```

```python
gamma = 0.5

lut = np.zeros((256, 1), dtype = 'uint8')

for i in range(len(lut)):
    lut[i][0] = 255 * pow((float(i)/255), (1.0 / gamma))

gamma_img = cv2.LUT(img, lut)

cv2.imwrite('gamma_img.jpg', gamma_img)
```

```python
files = os.listdir('./data/ants')
gamma = 0.6

while gamma < 1.2:
    lut = np.zeros((256, 1), dtype = "uint8")

    for i in range(len(lut)):
        lut[i][0] = 255 * pow((float(i)/255), (1.0 / gamma))
    for j in files:
        img = cv2.imread('./data/ants/' + j, 0)
        gamma_img = cv2.LUT(img, lut)
        cv2.imwrite('./ants_brightness' + str(j), gamma_img)
    gamma+=0.6
```
