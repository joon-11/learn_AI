**그레이스케일**

컬러 이미지에는 R G B에 0~255 까지의 픽셀값의 배열이 있다.
반면 그레이스케일 이밎에는 휘도(명암?)에 대해서만 0~255의 픽셀값이 배열에 들어가 있다.
컬러 이미지를 계산하려면 0~255 * 3을 해야하기때문에 속도가 오래걸리지만 그레이스케일을 쓰면 그렇지 않다.

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print(gray_img.shape)

plt.imshow(gray_img, cmap = 'gray')
plt.show()
```

```python
import os
import cv2
import numpy as np

dirs = ['ants', 'bees']
pixels = []
labels = []

for i, d in enumerate(dirs):
    files = os.listdir('./data/' + d)

    for f in files:
        img = cv2.imread('./data/' + d + '/' + f, 0)
        img = cv2.resize(img, (128, 128))
        img = np.array(img).flatten().tolist()
        pixels.append(img)

        labels.append(i)
```

**이진화 이미지**

픽셀값이 경계값보다 크면 백(255)로 주고 작으면 흑(0)으로 변환한다.

threshold() 를 이용해 경계값을 설정
threshold(이진화 하고싶은 픽셀값, 경계값, 최대값,thresholding type)

```python
ret, bin_img = cv2.threshold(gray_img, 128, 255, cv2.THRESH_BINARY)

plt.imshow(bin_img, cmap = 'gray')
plt.show()
```
